```python
import requests
from bs4 import BeautifulSoup
import time
import random
```

```python
url_base = 'http://fifaonline4.nexon.com/datacenter/rank_m?n4pageno='
nickname = []
for i in range(1,501):


    url = url_base + str(i)
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')


    list_nickname = soup.find_all('span', class_='name profile_pointer')
    
    for item in list_nickname:
        nickname.append(item.get_text())
    
    time.sleep(random.random())
    print("{}번째".format(i))

```
